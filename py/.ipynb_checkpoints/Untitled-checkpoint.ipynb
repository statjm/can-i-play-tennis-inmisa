```python
import requests
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
```

```python
yyyymm = "202202"
no_court = 1

url = "https://www.hanam.go.kr/www/selectMisaParkResveWeb.do?key=4062&yyyymm={}&misaParkCode=TS0{}&listType=D#n".format(yyyymm, no_court)

response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    # print(soup)

else: 
    print(response.status_code)
```

```python
tb_html = soup.find('table',{'class' : 'p-table'})
```

```python
tb_html = str(tb_html)
```

```python
table_df_list = pd.read_html(tb_html)
```

```python
table_df = table_df_list[0]
```

```python
table_df.head()
```

```python
table_df['회차'] = table_df['이용시간'].apply(lambda x: int(x.split("회")[0]))
```

```python
table_df['이용시간'] = table_df['이용시간'].apply(lambda x: x.split("회")[1])
```

```python
table_df['요일'] = table_df['일자'].apply(lambda x: x[-2])
table_df['일자'] = table_df['일자'].apply(lambda x: x[0:10])
```

```python
table_df['코트번호'] = no_court
```

```python
filtered_df = table_df[table_df.회차.isin([7, 8])]
filtered_user_df = filtered_df.copy()
```

```python
df_dict = {str(no_court)+'코트' : filtered_df}
```

```python
df_dict
```

```python
filtered_df.loc[:,str(no_court) + '코트'] = np.where(filtered_df['신청하기'] == "신청가능", "", np.where(pd.isnull(filtered_df['신청하기']), "예약불가\n{" + np.where(pd.isnull(filtered_df['예약단체']), "수수께기단체", filtered_df['예약단체']) + ")", filtered_df['신청하기']))
```
